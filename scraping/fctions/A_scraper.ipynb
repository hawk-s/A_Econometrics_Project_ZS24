```python
from B2_load_html_fctn import load_html_document


#soup of individual sites of sets (including the list of the corresponding sites)
soup_of_sets = load_html_document('ind_sets_sites2.html')

sets = {}  #empty dictionary

#find all <h1> tags since they contain the set names
for h1 in soup_of_sets.find_all('h1'):
    set_name = h1.text.strip()
    next_table = h1.find_next('table')  #find the table following the <h1> tag

    #get all links within this table
    links = [a['href'] for a in next_table.find_all('a', href=True)]
    sets[set_name] = links

    
#print(sets)

#'sets' now contains the dictionary with set names as keys and lists of links as values

#write_to_json('cards_links_wo.json',sets)
```

```python
from C1_json_writer import write_to_json

url_for_append_2 = 'https://www.pokemonprice.com'

#modify the dictionary to contain full URLs
for set_name, urls in sets.items():
    sets[set_name] = [url if url.startswith(url_for_append_2) else f"{url_for_append_2}{url}" for url in urls]

write_to_json('cards_linkss.json', sets) #create json containing individual cards links
```
